### Chosen Use Case:
Automated Supervision of Cashier Change Transactions in Supermarkets
 
 ### Problem Statement:
 In retail environments like supermarkets, ensuring cashiers provide the correct change is crucial for customer satisfaction and loss prevention. Manual supervision is impractical, and occasional mistakes, whether accidental or intentional, can lead to customer disputes and erode trust. An automated system is needed to monitor cash transactions, specifically the change given back, to detect discrepancies. The core challenge is to accurately identify and count the coins returned to the customer using visual information, often in real-time or near real-time, under potentially variable checkout counter conditions.
 
 ### Practical Use Case & Aim:
 The aim is to develop an automated supervision system that uses computer vision to monitor the coins placed on the counter or tray by a cashier as change. By segmenting and identifying the individual coins from camera footage overlooking the checkout area, the system can calculate the total value of the change given. This calculated value can then be compared against the expected change amount (derived from the point-of-sale system) to flag potential errors. This helps improve cashier accuracy, enhances customer experience by ensuring correct change, and provides a tool for management oversight.
